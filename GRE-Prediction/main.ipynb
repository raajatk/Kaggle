```python
import pandas as pd
```

```python
dataframe = pd.read_csv('./graduate-admissions/Admission_Predict.csv')
```

```python
dataframe.describe()
```

```python
dataframe['GREScore'] /= 316.807500
dataframe['TOEFLScore'] /= 107.410000
dataframe['UniversityRating'] /= 3.087500
dataframe['SOP'] /= 3.4
dataframe['LOR'] /= 3.4525
dataframe['CGPA'] /= 8.598925

dataframe.describe()
```

```python
dataframe.isnull().sum()
```

```python
target = dataframe['ChanceofAdmit']
dataframe = dataframe.drop(['ChanceofAdmit'],axis=1)
dataframe.describe()
```

```python
target.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataframe,target, test_size = 0.25, random_state = 0)
```

```python
X_train = X_train.drop(['SerialNo'],axis=1)
```

```python
X_train.describe()
```

```python
X_test = X_test.drop(['SerialNo'],axis=1)
```

```python
y_train.describe()
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
# accuracy = model.score(X_test,y_test)
```

```python
from sklearn import metrics
import numpy as np
print(metrics.mean_squared_error(y_test,y_pred))
print(metrics.mean_absolute_error(y_test,y_pred))
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.svm import SVR
model_svr = SVR(kernel='rbf')
model_svr.fit(X_train,y_train)
y_pred_svr = model_svr.predict(X_test)
```

```python
print(metrics.mean_squared_error(y_test,y_pred_svr))
print(metrics.mean_absolute_error(y_test,y_pred_svr))
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred_svr)))
```
