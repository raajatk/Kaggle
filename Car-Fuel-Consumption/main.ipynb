```python
# dataframe['temp_inside']
temp_inside = []
for i in dataframe['temp_inside']:
    if len(i.split(','))==2:
        #print(int(i.split(',')[0])+float(i.split(',')[1])*0.1)
        temp_inside.append(float(int(i.split(',')[0])+float(i.split(',')[1])*0.1))
    else:
        temp_inside.append(float(i))
```

```python
targ = []
for i in target:
    if len(i.split(','))==2:
        #print(int(i.split(',')[0])+float(i.split(',')[1])*0.1)
        targ.append(float(int(i.split(',')[0])+float(i.split(',')[1])*0.1))
    else:
        targ.append(float(i))
```

```python
target = pd.DataFrame(targ)
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(dataframe,target,test_size = 0.25, random_state = 0)
```

```python
y_test.describe()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
print(metrics.mean_squared_error(y_test,y_pred))
print(metrics.mean_absolute_error(y_test,y_pred))
```
